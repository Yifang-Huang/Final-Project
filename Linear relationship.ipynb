```python
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./Data_UK.csv')
corr = pearsonr(data['temperature'], data['R_t-estimate'])
corr = [np.round(c, 4) for c in corr]
print(corr)
X=data['temperature'].values.reshape(-1, 1)
pca=PCA(n_components='mle')
pca.fit(X)
X_train= pca.transform(X)
sns.set(style='whitegrid')
text = 'r=%s, p=%s' % (corr[0], corr[1])
ax = sns.regplot(x="temperature",y="R_t-estimate",ci= 95,data=data,marker="+",scatter_kws={"color": "#696969"}, line_kws={"color":"#4169E1"})
ax.text(33, 1.8, text, fontsize=15)
ax.set_title('Linear relationship_United Kingdom_Temperature_R')
```
